```python
import torch
import torch.nn as nn

class NonNegLoss(nn.Module):
    def __init__(self):
        super(NonNegLoss, self).__init__()

    def forward(self, tensor):
        return torch.sum(torch.relu(-tensor))

# Create a sample tensor (replace this with your actual tensor)
sample_tensor = torch.tensor([[-1, 2, -3, 4, -5]], dtype=torch.float32, requires_grad=True)

# Create an instance of NonNegLoss
non_neg_loss = NonNegLoss()

# Calculate the non-negativity loss
loss = non_neg_loss(sample_tensor)
print("Loss:")
print(loss)

# Backpropagation to compute gradients
loss.backward()

# Check the gradients of the tensor (gradients should be updated after backward)
print("Gradients of the tensor after backward:")
print(sample_tensor.grad)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate a 32,500 long sequence with a period longer than 325
sampling_rate = 1  # Hz
sequence_length = 32500
period_threshold = 325  # Hz

# Generate random data
sequence = np.random.rand(sequence_length)

# Perform an FFT on the sequence
y = np.fft.rfft(sequence)

# Create a high-pass filter in the frequency domain
filter_cutoff = period_threshold / sampling_rate
filter = np.where(np.abs(y) < filter_cutoff, 0, 1)

# Apply the filter in the frequency domain
y_filtered = y * filter

# Inverse transform the filtered sequence back to the time domain
filtered_sequence = np.fft.irfft(y_filtered)

# Plot the original and filtered sequences
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.plot(sequence, label='Original Sequence')
plt.title('Original Sequence')
plt.xlabel('Index')
plt.ylabel('Amplitude')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(filtered_sequence, label='Filtered Sequence')
plt.title('Filtered Sequence')
plt.xlabel('Index')
plt.ylabel('Amplitude')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
import numpy as np

# Assuming your array is named 'arr'
arr = np.random.rand(2, 2, 3)
print(arr)

# Find the maximum value along axis=2
max_val = np.max(arr, axis=2, keepdims=True)

# Find the minimum value along axis=2
min_val = np.min(arr, axis=2, keepdims=True)

# Normalize the array
normalized_arr = (arr - min_val) / (max_val - min_val)

# Print the resulting normalized array
print(normalized_arr)
```
